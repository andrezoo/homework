## Обычный перебор

```python
import numpy as np
import matplotlib.pyplot as plt 

n, times = [], []

with open('perebor_n.txt') as file:
    n = file.readlines()

with open('perebor_time.txt') as file:
    times = file.readlines()

n = np.array([int(en.replace('\n','')) for en in n])
times = np.array([int(it.replace('\n','')) for it in times])
plt.title('Зависимость времени t от N элементов для перебора')
plt.xlabel('N, размер массива')
plt.ylabel('t, время обработки')
plt.plot(n, times, linestyle='none', markersize=0.4,marker='o')
```

## Бинарный поиск

```python
import numpy as np
import matplotlib.pyplot as plt 

n, times = [], []

with open('binsearch_n.txt') as file:
    n = file.readlines()

with open('binsearch_time.txt') as file:
    times = file.readlines()

nn = []
tt = []

uplimit = 30
for i in range(0,len(times)):
    line = int(times[i].replace('\n',''))
    if line < uplimit:
        nn.append(int(n[i].replace('\n','')))
        tt.append(line)

plt.title('Зависимость времени t от N элементов для бинарного поиска')
plt.xlabel('N, размер массива')
plt.ylabel('t, время обработки')
plt.plot(nn, tt, linestyle='none', markersize=1.5,marker='o')
```
