```python
import pandas as pd 
import matplotlib.pyplot as plt
from scipy import optimize

def line_func(x, a, b):
    return a * x + b

d = pd.read_csv('perebor.txt')
#d = d.where(d['time'] < 1000).fillna(0)

params, params_covariance = optimize.curve_fit(line_func, d['n'], d['time'])

plt.scatter(d['n'], d['time'],s=0.75)
plt.plot(d['n'], line_func(d['n'],params[0],params[1]), c='green')
```
